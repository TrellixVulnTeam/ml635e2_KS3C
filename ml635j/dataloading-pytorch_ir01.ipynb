```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import random
from PIL import Image
from matplotlib import pyplot as plt
import xml.etree.ElementTree as ET

import torch 
import torchvision

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
```

```python

trainval = "../ir4/trainval"
#print(os.listdir(trainval))
mfiles = (os.listdir(trainval))

imgs = [image for image in sorted(os.listdir(trainval))
        if image[-4:]=='.png']
print(mfiles, '\n', imgs)
# Any results you write to the current directory are saved as output.
```

```python
imgpath = "../ir4/trainval"
# Files
images = os.listdir(imgpath)
print("Number of files: ", len(images))
```

# View Samples from Data

```python
# choose  random images to display
imgdisplay = random.choices(imgs, k=2)
print(imgdisplay)
fig = plt.figure(figsize=(25, 16))
for ii, img in enumerate(imgdisplay):
    ax = fig.add_subplot(8, 8, ii + 1, xticks=[], yticks=[])
    imgi = Image.open(os.path.join(imgpath, img))
    plt.imshow(imgi)
    
annot_filename = img[:-4] + '.xml'
#annot_file_path = os.path.join(trainval, annot_filename)
print(annot_filename)
        
### Getting Bounding Box of dog (ROI) in Image
anot = (os.path.join(imgpath, annot_filename))
tree = ET.parse(anot)
root = tree.getroot()
objects = root.findall('object')
for obj in objects:
    bndbox = obj.find('bndbox')
    xmin = int(bndbox.find('xmin').text)
    ymin = int(bndbox.find('ymin').text)
    xmax = int(bndbox.find('xmax').text)
    ymax = int(bndbox.find('ymax').text)
bbox = (xmin, ymin, xmax, ymax)
print("Bounding Box: ", bbox)

# crop image
imgc = imgi.crop(bbox)

# display crop image
fig = plt.figure(figsize=(8, 12))
plt.imshow(imgc)    
```
